# gradient descent in machine learning

```python
m = 0
c = 0 
lr = 0.001
interation = 1000
n = int(len(x))

for i in range(interation):
    y_pred = m*x + c
    derv_m = (-2/n)* sum(x*(y - y_pred))
    derv_c = (-2/n)* sum(y - y_pred)
    
    m = m - lr*derv_m
    c = c - lr*derv_c
print('m = ' , m)
print('c =' , c)

```

```python
y_pred = m*x + c
plt.scatter(x,y)
plt.plot(x,y)
plt.scatter(x,y_pred)
plt.plot(x,y_pred)
plt.show()
```
